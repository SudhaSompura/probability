### Bernoulli Random Variable

Parameters: $p \in [0,1]$ <br/>
Notation: $B(p)$ <br/>
Support: $ \{0,1\}$ <br/>
pmf: $ f(x) =\left\{
\begin{array}{ll}
      1-p & x=0 \\
      p & x=1 \\
\end{array} 
\right.$ <br/>
Expectation: $E[X] = p$ <br/>
Variance: $Var(X) = p(1-p)$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def bernoulli_pmf(p, ax):
      ax.axvline(0, 0, 1-p, label='probability of X=0')
      ax.axvline(1, 0, p, label='probability of X=1',c='r')
      ax.set_title("pmf")
      ax.set_xlim(-0.1,1.1)
      ax.grid()
      
      return
```

```python
def bernoulli_cdf(p, ax):
      ax.step([-0.5, 0,1, 2], [0.0, 0.0, 1-p, 1])
      ax.set_xlim(-0.5, 1.5)
      ax.set_title('cdf')
      ax.grid()
      return
```

```python
def main():
      sub_figs = 4
      x = 0
      
      for p in np.linspace(0.1,0.9,sub_figs):
            fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (6,3), constrained_layout=True)
            plt.setp(axes, xticks=[0, 1])
            bernoulli_pmf(p, axes[0])
            bernoulli_cdf(p, axes[1])
            fig.suptitle("Bernoulli distribution with B(p=" + str(round(p, 2)) + ")")  
            plt.show()
            x += 1
```

```python
main()
```
